```python
#PROBLEM 4
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# 1. A * A^-1 = I
invA = np.linalg.inv(A)
prove1 = A @ invA  # should be identity

# 2. AB != BA
AB = A @ B
BA = B @ A

# 3. (AB)^T = B^T A^T
lhs = (AB).T
rhs = B.T @ A.T

prove1, AB, BA, lhs, rhs
```
