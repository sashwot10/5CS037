Comparison of Composite Score Ranking vs HDI Ranking (South Asia)

The Composite Score and HDI rankings show a strong positive relationship, indicating that countries with higher HDI values generally also achieve higher Composite Scores. This is expected because both measures reflect core development dimensions such as health and income.

However, some ranking differences are observed. For example, Sri Lanka has the highest HDI among South Asian countries, but Maldives ranks highest in Composite Score. This difference occurs because the Composite Score gives greater weight to GNI per capita and life expectancy, where Maldives performs exceptionally well due to its higher income levels.

Conversely, countries such as Bhutan and Bangladesh show relatively closer rankings between HDI and Composite Score, suggesting balanced performance across both health and income indicators. Nepal and Pakistan remain lower in both rankings, reflecting comparatively lower life expectancy and income levels.

Overall, while HDI and Composite Score are positively correlated, they are not identical. HDI incorporates education more strongly, whereas the Composite Score focuses only on life expectancy and income. As a result, countries with strong economic performance may rank higher in Composite Score even if their HDI is slightly lower, highlighting how different indicator weightings can influence development rankings.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/Data/HDI SouthAsia.csv")


#Detect outliers in HDI and GNI per Capita using the 1.5 × IQR rule.
def iqr_outliers(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return (series < lower) | (series > upper)

df["hdi_outlier"] = iqr_outliers(df["hdi"])
df["gni_outlier"] = iqr_outliers(df["gross_inc_percap"])
df["outlier"] = df["hdi_outlier"] | df["gni_outlier"]

plt.figure()
plt.scatter(df.loc[~df["outlier"], "gross_inc_percap"], df.loc[~df["outlier"], "hdi"])
plt.scatter(df.loc[df["outlier"], "gross_inc_percap"], df.loc[df["outlier"], "hdi"])
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("HDI vs GNI per Capita (Outliers Highlighted)")
plt.show()
outliers_df = df[df["outlier"]]
outlier_countries = outliers_df["country"].unique()
print(outlier_countries)
```

Sri Lanka and Maldives stand out as outliers because their GNI per Capita and HDI are much higher than the rest of South Asia. Maldives benefits from a small population and tourism-driven economy, while Sri Lanka has higher development indicators. Other countries cluster together, so they are not statistically outliers.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load datasets
df_south_asia = pd.read_csv("/content/drive/MyDrive/Data/HDI SouthAsia_2020_2022.csv")
df_middle_east = pd.read_csv("/content/drive/MyDrive/Data/HDI MiddleEast_2020_2022.csv")

# Compute mean HDI per country
mean_hdi_sa = df_south_asia.groupby("country")["hdi"].mean()
mean_hdi_me = df_middle_east.groupby("country")["hdi"].mean()

# Top 3 performers
top3_sa = mean_hdi_sa.sort_values(ascending=False).head(3).reset_index()
top3_me = mean_hdi_me.sort_values(ascending=False).head(3).reset_index()

# Bottom 3 performers
bottom3_sa = mean_hdi_sa.sort_values(ascending=True).head(3).reset_index()
bottom3_me = mean_hdi_me.sort_values(ascending=True).head(3).reset_index()

# Combine top and bottom performers for plotting
combined_sa = pd.concat([top3_sa, bottom3_sa], ignore_index=True)
combined_me = pd.concat([top3_me, bottom3_me], ignore_index=True)

# Positions for bars
x = np.arange(len(combined_sa))  # 6 positions (top3 + bottom3)
width = 0.35

# Create combined labels for x-axis: "SA country\nME country"
x_labels = [f"{sa}\n{me}" for sa, me in zip(combined_sa["country"], combined_me["country"])]

# Plot vertical grouped bar chart
plt.figure(figsize=(12,6))
plt.bar(x - width/2, combined_sa["hdi"], width, color="skyblue", label="South Asia")
plt.bar(x + width/2, combined_me["hdi"], width, color="salmon", label="Middle East")
plt.xticks(x, x_labels)
plt.ylabel("Average HDI (2020–2022)")
plt.title("Top 3 and Bottom 3 HDI Performers: South Asia vs Middle East")
plt.legend()
plt.show()
```

```python
import pandas as pd

# Load datasets (if not already loaded)
df_south_asia = pd.read_csv("/content/drive/MyDrive/Data/HDI SouthAsia_2020_2022.csv")
df_middle_east = pd.read_csv("/content/drive/MyDrive/Data/HDI MiddleEast_2020_2022.csv")

# Compute HDI range (max - min)
range_sa = df_south_asia["hdi"].max() - df_south_asia["hdi"].min()
range_me = df_middle_east["hdi"].max() - df_middle_east["hdi"].min()

# Compute coefficient of variation (CV = std / mean)
cv_sa = df_south_asia["hdi"].std() / df_south_asia["hdi"].mean()
cv_me = df_middle_east["hdi"].std() / df_middle_east["hdi"].mean()

print(f"South Asia HDI Range: {range_sa:.3f}")
print(f"Middle East HDI Range: {range_me:.3f}")
print(f"South Asia HDI CV: {cv_sa:.3f}")
print(f"Middle East HDI CV: {cv_me:.3f}")

# Identify which region exhibits more variation
if cv_sa > cv_me:
    print("South Asia exhibits more relative variation in HDI.")
else:
    print("Middle East exhibits more relative variation in HDI.")
```

The HDI range and coefficient of variation show that the Middle East exhibits more variation in human development than South Asia. Specifically, the HDI range is 0.513 for the Middle East versus 0.321 for South Asia, and the CV is 0.179 versus 0.154. This indicates that HDI outcomes are more unequal among Middle Eastern countries, while South Asia shows relatively less disparity.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets if not already loaded
df_south_asia = pd.read_csv("/content/drive/MyDrive/Data/HDI SouthAsia_2020_2022.csv")
df_middle_east = pd.read_csv("/content/drive/MyDrive/Data/HDI MiddleEast_2020_2022.csv")

# Metrics to correlate with HDI
metrics = ["gender_development", "life_expectancy"]

# Function to compute correlations and plot scatter with trendline
def correlation_analysis(df, region_name):
    print(f"\n--- {region_name} ---")
    for metric in metrics:
        corr = df["hdi"].corr(df[metric])
        print(f"Correlation between HDI and {metric.replace('_',' ').title()}: {corr:.3f}")

        # Scatter plot with trendline
        plt.figure(figsize=(6,4))
        sns.regplot(x=metric, y="hdi", data=df, scatter_kws={'color':'skyblue'}, line_kws={'color':'red'})
        plt.xlabel(metric.replace("_"," ").title())
        plt.ylabel("HDI")
        plt.title(f"{region_name}: HDI vs {metric.replace('_',' ').title()}")
        plt.show()

# Run analysis for both regions
correlation_analysis(df_south_asia, "South Asia")
correlation_analysis(df_middle_east, "Middle East")
```

In South Asia, HDI shows a strong positive correlation with both Life Expectancy (0.939) and Gender Development Index (0.874), indicating that higher life expectancy and better gender equality are associated with higher HDI. In the Middle East, HDI is also strongly positively correlated with Gender Development (0.936) and Life Expectancy (0.933). Overall, Life Expectancy is slightly more strongly related to HDI in South Asia, while in the Middle East, both metrics are almost equally strongly correlated. This shows that improvements in health and gender equality consistently contribute to higher human development in both regions.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
df_south_asia = pd.read_csv("/content/drive/MyDrive/Data/HDI SouthAsia_2020_2022.csv")
df_middle_east = pd.read_csv("/content/drive/MyDrive/Data/HDI MiddleEast_2020_2022.csv")

# Function to detect outliers using 1.5*IQR
def detect_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    return outliers

# Detect outliers for HDI and GNI per Capita
outliers_sa_hdi = detect_outliers(df_south_asia, "hdi")
outliers_sa_gni = detect_outliers(df_south_asia, "gross_inc_percap")

outliers_me_hdi = detect_outliers(df_middle_east, "hdi")
outliers_me_gni = detect_outliers(df_middle_east, "gross_inc_percap")

print("South Asia HDI Outliers:")
print(outliers_sa_hdi[["country", "year", "hdi"]])
print("\nSouth Asia GNI Outliers:")
print(outliers_sa_gni[["country", "year", "gross_inc_percap"]])

print("\nMiddle East HDI Outliers:")
print(outliers_me_hdi[["country", "year", "hdi"]])
print("\nMiddle East GNI Outliers:")
print(outliers_me_gni[["country", "year", "gross_inc_percap"]])

# Scatter plot function
def scatter_outliers(df, x_col, y_col, outliers, title):
    plt.figure(figsize=(7,5))
    plt.scatter(df[x_col], df[y_col], label="Normal", color="skyblue")
    plt.scatter(outliers[x_col], outliers[y_col], label="Outliers", color="red")
    plt.xlabel(x_col.replace("_"," ").title())
    plt.ylabel(y_col.replace("_"," ").title())
    plt.title(title)
    plt.legend()
    plt.show()

# Plot HDI vs GNI for each region with outliers highlighted
scatter_outliers(df_south_asia, "gross_inc_percap", "hdi", outliers_sa_hdi, "South Asia: HDI vs GNI per Capita (HDI Outliers)")
scatter_outliers(df_middle_east, "gross_inc_percap", "hdi", outliers_me_hdi, "Middle East: HDI vs GNI per Capita (HDI Outliers)")
```


  In South Asia, there are no HDI or GNI per Capita outliers, meaning the countries are relatively clustered in terms of development and income.


  In the Middle East, Yemen is an HDI outlier for all three years (2020–2022), with much lower HDI than its regional peers.
  This highlights Yemen’s severe human development challenges, likely due to ongoing conflict, economic instability, and limited access to health and education.


  There are no GNI per Capita outliers in either region, indicating income levels are relatively consistent compared to the spread in HDI.
