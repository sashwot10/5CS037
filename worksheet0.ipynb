```python
def sum_nested_list(nested_list):
    """
    Calculate the sum of all numbers in a nested list.
    This function takes a list that may contain integers and other nested lists.
    It recursively traverses the list and sums all the integers, no matter how deeply nested they are.
    """
    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)
        else:
            total += element
    return total
```

```python
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
result = sum_nested_list(nested_list)
print("Total Sum:", result)
```

```python
def generate_permutations(s):
    result = []
    used = [False] * len(s)
    s = list(s)  # convert string to list for easier handling

    def backtrack(path):
        if len(path) == len(s):
            result.append("".join(path))
            return

        for i in range(len(s)):
            if used[i]:
                continue

            # Avoid duplicates for repeated characters
            if i > 0 and s[i] == s[i-1] and not used[i-1]:
                continue

            used[i] = True
            backtrack(path + [s[i]])
            used[i] = False

    s.sort()  # sorting handles duplicate characters properly
    backtrack([])
    return result
```

```python
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}
```

```python
def calculate_directory_size(directory):
    total_size = 0

    for name, value in directory.items():
        if isinstance(value, dict):    # It's a subdirectory
            total_size += calculate_directory_size(value)
        else:                          # It's a file (size in KB)
            total_size += value

    return total_size
```

```python
size = calculate_directory_size(directory_structure)
print("Total Directory Size:", size, "KB")
```

```python
def min_coins(coins, amount):
    """
    Finds the minimum number of coins needed to make up a given amount using dynamic programming.
    If it's not possible, returns -1.
    """
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1
```

```python
print(min_coins([1, 2, 5], 11))  # Expected: 3 (5 + 5 + 1)
print(min_coins([2], 3))        # Expected: -1
```

```python
def longest_common_subsequence(s1, s2):
    n = len(s1)
    m = len(s2)

    dp = [[0] * (m + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = 1 + dp[i - 1][j - 1]
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[n][m]
```

```python
print(longest_common_subsequence("abcde", "ace"))  # Expected: 3
print(longest_common_subsequence("aggtab", "gxtxayb"))  # Expected: 4 ("gtab")
```

```python
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(values[i - 1] + dp[i - 1][w - weights[i - 1]], dp[i - 1][w])
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacity]
```

```python
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7

print(knapsack(weights, values, capacity))  # Expected: 9
```
